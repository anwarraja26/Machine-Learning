```python
years=range(2000,2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges= [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896]
```

## Plot multiple lines

```python
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('Year')
plt.ylabel('Yield')
```

## Line Markers

```python
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')

plt.xlabel('Years')
plt.ylabel('yields')

plt.title('Crop Yield in India')
plt.legend(['Apples','Oranges'])
```

### Styling lines and markers

### The plt.plot function supports many arguments for styling lines and markers:
*  
color or c: set the color of the line (supported color ,s*  )
linestyle or ls: choose between a solid or dashed li*  ne
linewidth or lw: set the width of a l*  ine
markersize or ms: set the size of mar*  kers
markeredgecolor or mec: set the edge color for ma*  rkers
markeredgewidth or mew: set the edge width for m*  arkers
markerfacecolor or mfc: set the fill color for *  markers
alpha: opacity of the plot

```python
plt.plot(years,apples,marker='s',c='g',ls="-",lw=2,ms=8,mew=3,mec='blue',mfc="red",alpha=.9)
plt.plot(years,oranges,marker="o",c="red",ls="-",lw=4,ms=8,mew=2,mec="black",mfc="yellow",alpha=.8)

plt.xlabel('Years')
plt.ylabel('yields')

plt.title('Crop Yield in India')
plt.legend(['Apples','Oranges'])
```

## HISTOGRAM

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.columns
```

```python
plt.title("Distribution of sepal length")
plt.hist(df.sepal_length)
```

```python
import numpy as np
plt.title("histogram with bins size")
plt.hist(df.sepal_width,bins=np.arange(2,5,0.5))
```

```python
df.species.unique()
```

## Multiple histogram

```python
setosa_df=df[df.species=="setosa"]
versicolor_df=df[df.species=="versicolor"]
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.5,bins=np.arange(2,5,0.2))
plt.hist(versicolor_df.sepal_width,alpha=0.5,bins=np.arange(2,5,0.2))
```

## Multiple histogram in stack of one another

```python
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width],bins=np.arange(2,5,0.2),stacked=True)
```
