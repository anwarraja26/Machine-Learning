```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Example DataFrame
data = {'Airline': ['AirAsia', 'Indigo', 'SpiceJet'], 
        'Source': ['Delhi', 'Mumbai', 'Kolkata'], 
        'Destination': ['Chennai', 'Delhi', 'Mumbai']}
df = pd.DataFrame(data)

# Initialize OneHotEncoder
encoder = OneHotEncoder()

# Apply OneHotEncoder to the selected columns
encoded_data = encoder.fit_transform(df[['Airline', 'Source', 'Destination']])

# Convert the sparse matrix to a DataFrame with meaningful column names
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['Airline', 'Source', 'Destination']))

# Combine with original DataFrame or replace columns
df = pd.concat([df, encoded_df], axis=1)

print(df)
```
