# PIPE LINE

### The term pipeline refers to the combiation of both the estimators like linear_model and the transformer like standard scaler,minmax scaler together into a single unifying object .It helps to prevent you from data leakage 

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# create the pipeline we can put the estimator more than one and the transformer can be in last or first the transformer dont work if there is estimator next transformer
pipe=make_pipeline(
    StandardScaler(), # transformer
    MinMaxScaler(), #transformer
    LogisticRegression() #estimator
)
# but we cant pass like below bcoz we cant use the estimator in between two transformer
# make_pipeline(
#     StandardScaler(), #transformer
#     LogisticRegression(),#estimator
#     MinMaxScaler(),#transformer
# )
# #
```

```python
X,y=load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.33)
```

```python
pipe.fit(X_train,y_train) #fit the whole pipeline
```

```python
# we can now use it like any other estimator
accuracy_score(pipe.predict(X_test),y_test)
```

## Automated Parameter search

```python
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from scipy.stats import randint

X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# define the parameter space that will be searched over
param_distributions = {'n_estimators': randint(1, 5),
                       'max_depth': randint(5, 10)}

# now create a searchCV object and fit it to the data
search = RandomizedSearchCV(estimator=RandomForestRegressor(random_state=0),
                             n_iter=5,
                             param_distributions=param_distributions,
                             random_state=0)

```

```python
search.fit(X_train,y_train)
```

```python
search.best_params_
```

```python
search.score(X_test,y_test)
```

```python
search.best_estimator_.score(X_test,y_test)
```
