# it is used to merge the two data frames 
# it will merge the tables on the country code column based on the left table like left join

```python
final=pd.merge(df,df_country,on='Country Code',how='left') 
```

```python
final.columns
```

```python
final.head(2)
```

```python
final.Country.value_counts()
```

```python
country_names=final.Country.value_counts() .index
```

```python
country_value=final.Country.value_counts().values
```

```python
import matplotlib.pyplot as plt
plt.pie(country_value,labels=country_names)
```

```python
# Here we need to find the top 3 city only and only want two points in percentage
import matplotlib.pyplot as plt
plt.pie(country_value[:3],labels=country_names[:3],autopct='%1.2f%%')
```

# Here we are goind to group the rating related columns 

```python
final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # it help to increase the size of the picture
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
# in the above we not used the rating color for the bars based on the color the below chart is made 
# using the hue keyword
import matplotlib
# matplotlib.rcParams['figure.figsize']=(12,6) # it help to increase the size of the picture
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings)
```

```python
# in the above the right side box name fo the color and the colr is not the same so 
#to solve this use palette
# in the above we not used the rating color for the bars based on the color the below chart is made 
# using the hue keyword
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # it help to increase the size of the picture
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',palette=['blue','red','orange','yellow','green','green'],data=ratings)
```

Observation 
    1.Not rated count is ver high
    2. maximum number of rating is between 2.5 to 3.4


```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'],hue='Rating color')
```
